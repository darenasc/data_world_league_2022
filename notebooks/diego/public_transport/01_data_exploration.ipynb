```python
import pandas as pd
from pandas_profiling import ProfileReport
from pathlib import Path
```

```python
output_folder = Path("../../../output/diego/cascais")
cascais_path = Path("data/cascais/")
```

```python
dfs = {}
# Generating exploratory analysis for the input data
for item in cascais_path.glob('*.txt'):
    df = pd.read_csv(item, sep=',')
    df.name = item.stem
    profile = ProfileReport(df, title=f"{item.stem}")
    profile.to_file(output_folder / f"{item.stem}.html")
    dfs[item.stem] = df
```

- `stop_times` is a relevant dataset.
- `trips` has a `shape_id` column that joins with the `shapes` file.
- `stops` contains the coodinates of the bus stops.
